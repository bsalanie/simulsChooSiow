```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set()

%matplotlib inline

Phi_comps = np.load("../results/SmallComp/Phi_comp_sim200.npy")
Phi_comp_true = 0.1

w = 8
h = 5

plt.figure(figsize=(w, h), dpi=300)
sns.distplot(Phi_comps, hist=False,color='red', label='Phi_mod')
plt.axvline(Phi_comp_true, color='red',  linestyle='--', linewidth=1)
plt.text(x=0.45,y=1.05, s="over 200 samples of 400M/400W:")
plt.text(x=1,y=0.95, s="mean   = 0.103")
plt.text(x=1,y=0.9, s="stderr  = 0.321")

plt.title('Estimates of Phi_mod')
plt.legend()

plt.savefig("../results/SmallComp/Phi_mod_estimates.png")

```
